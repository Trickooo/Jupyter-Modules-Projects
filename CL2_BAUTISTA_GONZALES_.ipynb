# 1. Grade Sorter App

<b>Description:</b>
<br>You are responsible for writing a program that will collect four grades from a user. Your program will then sort these grades from highest to lowest. Then, your program will simulate dropping the lowest two grades the user entered. Lastly, it will comment on the users highest grade.

<b>Example Output:</b>
<br>Welcome to the Grade Sorter App
<br>What is your first grade (0-100): 82
<br>What is your second grade (0-100): 95
<br>What is your third grade (0-100): 100
<br>What is your fourth grade (0-100): 61

<br>Your grades are: [82, 95, 100, 61]

<br>Your grades from highest to lowest are: [100, 95, 82, 61]

<br>The lowest two grades will now be dropped.
<br>Removed grade: 61
<br>Removed grade: 82

<br>Your remaining grades are: [100, 95]
<br>Nice work! Your highest grade is a 100.

```python
print("Welcome to joy and Patrick's Grade Sorter App")
print("An app that will help you list your grades, sort and see your highest grade!")

grades = []
grade = int(input("\nWhat is your first grade (0-100): "))
grades.append(grade)
grade = int(input("\nWhat is your second grade (0-100): "))
grades.append(grade)
grade = int(input("\nWhat is your third grade (0-100): "))
grades.append(grade)
grade = int(input("\nWhat is your fourth grade (0-100): "))
grades.append(grade)
print("\nYour grades are: " + str(grades))
print("\nYour grades from highest to lowest are: ", end= " ")
print(sorted(grades, reverse = True))
print("\nThe lowest two grade will now be dropped.")
grades.sort(reverse=True)
removeGrade = grades.pop()
print("We removed grade: " + str(removeGrade))
removeGrade = grades.pop()
print("We removed grade: " + str(removeGrade))
print("\nThe remaining grades on your listare: " + str(grades))
print("\nCongratiolations! You did great! Your highest grade is: " + str(grades[0]))
```

# 2. Different Types of Lists Program

```python
num_strings = ["15","100","55","42"]
num_ints = [15, 100, 55, 42]
num_floats = [2.2, 5.0, 1.245, 0.142857]
num_lists = [[1, 2, 3],[4, 5, 6],[7, 8, 9]]

print("The variable num_strings is a", end= " ")
print(type(num_strings))
print("it contains the elements: ", end= " ")
print(num_strings)
print("The element " + str(num_strings[0]) + " is a", end= " ")
print(type(num_strings[0]))

print("\nThe variable num_ints is a", end= " ")
print(type(num_ints))
print("it contains the elements: ", end= " ")
print(num_ints)
print("The element " + str(num_ints[0]) + " is a", end= " ")
print(type(num_ints[0]))

print("\nThe variable num_floats is a", end= " ")
print(type(num_floats))
print("it contains the elements: ", end= " ")
print(num_floats)
print("The element " + str(num_floats[0]) + " is a", end= " ")
print(type(num_floats[0]))

print("\nThe variable num_lists is a", end= " ")
print(type(num_lists))
print("it contains the elements: ", end= " ")
print(num_lists)
print("The element " + str(num_lists[0]) + " is a", end= " ")
print(type(num_lists[0]))

print("\nNow sorting the num_strings and num_ints...")
print("Sorted num_strings: ", end= " ")
print(sorted(num_strings))
print("Sorted num_ints: ", end= " ")
print(sorted(num_ints))

print("\nThe main difference between sorting a string to sorting a integer")
print("is that sorting the string would only sort the first number,")
print("so despite being higher, if your number starts low, it would be the first one.")
print("While the integers allows the lists to be sorted based on the number.")
```
